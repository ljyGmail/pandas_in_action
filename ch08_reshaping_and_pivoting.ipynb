## 8.2 Creating a pivot table from a DataFrame

```python
import pandas as pd
```

```python
pd.read_csv("data/ch08/sales_by_employee.csv").head()
```

```python
sales = pd.read_csv(
    "data/ch08/sales_by_employee.csv", parse_dates=["Date"]
)

sales.tail()
```

### 8.2.1 The pivot_table method

```python
pd.pivot_table(data=sales, index=["Date"], values=["Revenue", "Expenses"])
```

```python
# the default aggfunc is mean
pd.pivot_table(sales, index="Date", values=["Revenue", "Expenses"], aggfunc="mean")
```

```python
pd.pivot_table(sales, index="Date", values=["Revenue", "Expenses"], aggfunc="sum")
```

```python
# only care about summing the values in the Revenue column
pd.pivot_table(sales, index="Date", values="Revenue", aggfunc="sum")
```

```python
pd.pivot_table(sales, index="Date", columns="Name", values="Revenue", aggfunc="sum")
```

```python
pd.pivot_table(sales, 
               index="Date", 
               columns="Name", 
               values="Revenue", 
               aggfunc="sum",
               fill_value=0
)
```

```python
pd.pivot_table(sales,
              index="Date",
              columns="Name",
              values="Revenue",
              aggfunc="sum",
              fill_value=0,
              margins=True
)
```

```python
pd.pivot_table(sales,
              index="Date",
              columns="Name",
              values="Revenue",
              aggfunc="sum",
              fill_value=0,
              margins=True,
              margins_name="Total"
)
```

### 8.2.2 Additional options for pivot tables

```python
# number of business deals closed per day
pd.pivot_table(sales,
              index="Date",
              columns="Name",
              values="Revenue",
              aggfunc="count"
)
```

```python
# pass a list of aggregation functions
pd.pivot_table(sales,
              index="Date",
              columns="Name",
              values="Revenue",
              aggfunc=["sum", "count"],
              fill_value=0)
```

```python
pd.pivot_table(sales,
              index="Date",
              columns="Name",
              values=["Revenue", "Expenses"],
              fill_value=0,
              aggfunc={"Revenue": "min", "Expenses": "max"})
```

```python
# stack multiple groupings on a single axis
pd.pivot_table(sales,
              index=["Name", "Date"],
              values="Revenue",
              aggfunc="sum").head(10)
```

```python
# switch the order of strings in the index
pd.pivot_table(sales,
              index=["Date", "Name"],
              values="Revenue",
              aggfunc="sum").head(10)
```

## 8.4 Melting a data set

```python
sales.head(1)
```

```python
video_game_sales = pd.read_csv("data/ch08/video_game_sales.csv")
video_game_sales.head()
```

```python
video_game_sales.head(1)
```

```python
pd.melt(video_game_sales, id_vars="Name", value_vars="NA").head()
```

```python
regional_sales_columns = ["NA", "EU", "JP", "Other"]

pd.melt(video_game_sales, id_vars="Name", value_vars=regional_sales_columns)
```

```python
# customize the melted DataFrame's column names
video_game_sales_by_region = pd.melt(video_game_sales,
       id_vars="Name",
       value_vars=regional_sales_columns,
       var_name="Region",
       value_name="Sales"
)

video_game_sales_by_region.head()
```

```python
# Narrow data is easier to aggregate than wide data.
pd.pivot_table(video_game_sales_by_region, index="Name", values="Sales", aggfunc="sum").head()
```

## 8.6 Coding challenge

### 8.6.1 Problems

```python
cars = pd.read_csv("data/ch08/used_cars.csv")
cars.head()
```

```python
min_wage = pd.read_csv("data/ch08/minimum_wage.csv")
min_wage.head()
```

### 8.6.2 Solutions

```python
# 1. Aggregate the sum of car price in cars. Group the results by fuel type on the row axis.
pd.pivot_table(cars, values="Price", index="Fuel", aggfunc="sum")
```

```python
# 2. Aggregate the count of cars in cars. Group the results by manufacturer on the index axis
# and transmission type on the column axis. Show the subtotals for both the rows and columns.
pd.pivot_table(cars,
              values="Price",
              index="Manufacturer",
              columns="Transmission",
              aggfunc="count",
              margins=True
).tail()
```

```python
# 3. Aggregate the average of car prices in cars. Group the results by year and fuel type
# on the index axis and transmission type on the column axis.
report = pd.pivot_table(cars,
              values="Price",
              index=["Year", "Fuel"],
              columns=["Transmission"],
              aggfunc="mean"
)

report
```

```python
# 4. Given a DataFrame from the preceding challenge, move the transmission level from
# the column axis to the row axis.
report.stack()
```

```python
# 5. Convert the min_wage from wide format to narrow format. In other words, move the data
# from the eight year columns (2010-2017) to a single column.
year_columns = [
    "2010", "2011", "2012", "2013",
    "2014", "2015", "2016", "2017",
]

min_wage.melt(id_vars="State", value_vars=year_columns)
```

```python
# By default, pandas melts data from all columns except the one we pass to the id_vars parameter:
min_wage.melt(id_vars="State")
```

```python
# customize the column names
min_wage.melt(
    id_vars="State", var_name="Year", value_name="Wage"
)
```
